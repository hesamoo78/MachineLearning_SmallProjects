<p style="text-align: left; line-height:200%; font-family:vazir; font-size:medium;">
  <font face="vazirmatn" size="3">
    As you can see, except for the TotalTime, FuelEconomy, Toll, and Comments columns, all columns contain non-NaN values.<br/>
    The number of NaN values in the Comments column far exceeds its non-NaN values. Therefore, remove this column from the dataset.
  </font>
</p>


````python
# drop comments column
df.drop(columns = 'Comments', inplace = True)
df
````


<p style="text-align: left; line-height:200%; font-family:vazir; font-size:medium;">
  <font face="vazirmatn" size="3">
    Fill the missing values in the FuelEconomy and AvgMovingSpeed columns with the mean of those columns. (Round the calculated means to two decimal places using the round function.)
  </font>
</p>



````python
# fill NaN values in the AvgMovingSpeed and FuelEconomy
mean = df['FuelEconomy'].mean().round(2)
df.loc[:, 'FuelEconomy'] = df['FuelEconomy'].fillna(mean)
mean = df['AvgMovingSpeed'].mean().round(2)
df.loc[:, 'AvgMovingSpeed'] = df['AvgMovingSpeed'].fillna(mean)
df
````

<p style="text-align: left; line-height:200%; font-family:vazir; font-size:medium;">
  <font face="vazirmatn" size="3">
    One way to fill missing values is to predict them using machine learning models. To that end, execute the code cells below to locate the NaN values in the TotalTime and Toll columns. However, don't expect to understand this code right away.<br/>
    When you read the related lesson later in the course, return to this section of the lesson.
  </font>
</p>


````python
automl_reg = AutoML()
x_train, x_test = df.dropna(subset=['TotalTime']), df[df.TotalTime.isna()]
y_train = df[~df.TotalTime.isna()].TotalTime
# removing the columns with null values
x_train = x_train.drop(columns=['TotalTime','Toll'])
x_test = x_test.drop(columns=['TotalTime','Toll'])
# run the model
automl_reg.fit(x_train, y_train, estimator_list=["rf"], task="regression", verbose=False)
y_pred = automl_reg.predict(x_test)
````

````python
df.loc[df.TotalTime.isna(), 'TotalTime'] = y_pred
df.info()
````

````python
automl_clf = AutoML()
x_train, x_test = df.dropna(subset=['Toll']), df[df.Toll.isna()]
y_train = df[~df.Toll.isna()].Toll
# removing the columns with null values
x_train = x_train.drop(columns=['Toll'])
x_test = x_test.drop(columns=['Toll'])
# run the model
automl_clf.fit(x_train, y_train, task="classification", estimator_list=["rf"], verbose=False)
y_pred = automl_clf.predict(x_test)
````

````python
df.loc[df.Toll.isna(), 'Toll'] = y_pred
df.info()
````

<p style="text-align: left; line-height:200%; font-family:vazir; font-size:medium;">
  <font face="vazirmatn" size="3">
    As you can see in the boxplot for MaxSpeed, there are many outliers in this column (even negative values for speed!). Therefore, these values must be identified and removed.<br/>
    Below, using quartiles and calculating the IQR, identify the outliers in the MaxSpeed column and store them in the list <code>outlier_speed</code> in sorted order (from lowest to highest). Then replace the outlier values in MaxSpeed with the mean of that column, rounded to two decimal places using the <code>round</code> function. (Note that outliers should not influence the mean; calculate the mean excluding them.)
  </font>
</p>


<p style="text-align: left; line-height:200%; font-family:vazir; font-size:medium;">
  <font face="vazirmatn" size="3">
    If you print <code>outlier</code>, your output should look like this:<br/>
  </font>
</p>

```python
[-138.0, -120.0, -32.5, 112.2, 114.4, ...]

```

</font>
</p>

````python

# calculating quantiles to detect outliers
Q1, Q2, Q3 = df.MaxSpeed.quantile([0.25, 0.5, 0.75])
IQR = Q3 - Q1
outlier_speed = df['MaxSpeed'][
    (df['MaxSpeed'] > Q3 + 1.5 * IQR) |
    (df['MaxSpeed'] < Q1 - 1.5 * IQR)
].tolist()
outlier_speed.sort()
outlier_speed

````

````python
# replace outliers with avg of MaxSpeed 
mask = (
    (df['MaxSpeed'] > Q3 + 1.5 * IQR) |
    (df['MaxSpeed'] < Q1 - 1.5 * IQR)
)


mean_overall = df['MaxSpeed'].mean()

mean_clean = df.loc[~mask, 'MaxSpeed'].mean()


df.loc[mask, 'MaxSpeed'] = mean_clean
df
````
