
<h2 align="left" style="line-height:200%; font-family:vazir; color:#0099cc">
  <font face="vazirmatn" color="#0099cc">
    Part One
  </font>
</h2>

<p dir="ltr" style="direction: ltr; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
  <font face="vazirmatn" size="3">
    As mentioned before, the categorical columns in this DataFrame contain missing values. As the first part of this question, we ask you to:
  </font>
</p>

<ol dir="ltr" style="direction: ltr; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
  <font face="vazirmatn" size="3">
    <li>
      Gather the names of all categorical columns of the DataFrame into a list named <code>categorical_cols</code>. The target variable should not be included in this list.
    </li>
    <li>
      Then, for each column listed in <code>categorical_cols</code>, replace its missing values with the column’s most frequent value. Store the resulting DataFrame, which has no missing values, back into <code>df</code>. Thus, after completing part one, <code>df</code> should contain no missing values.
    </li>
  </font>
</ol>


```python
# TO-DO: fill `categorical_cols` and fill na values in categorical columns of `df` dataframe.
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'is_telephone_contact', 'month', 'day_of_week', 'poutcome']
for col in categorical_cols:
    most_freq = df[col].mode()[0]
    df[col] = df[col].fillna(most_freq)


df.head()
```

<p dir="ltr" style="direction: ltr; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
  <font face="vazirmatn" size="3">
    In part three, we want you to create a DataFrame that includes only the binary columns of the <code>df</code> DataFrame. Then convert all values in this new DataFrame to numeric.
    <br>
    Using <code>binary_cols</code>, you can construct the <code>binary_df</code> DataFrame.
    Essentially, <code>binary_df</code> is the portion of <code>df</code> that contains only its binary columns—meaning there are no continuous, nominal, or ordinal columns in <code>binary_df</code>. Note that all values in this DataFrame must be numeric (use <code>1</code> instead of <code>yes</code> and <code>0</code> instead of <code>no</code>).

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

binary_df=pd.DataFrame()
for i in binary_cols:
    binary_df[i]=label_encoder.fit_transform(df[i])
binary_df.head()
```

    <br>
    Using <code>ordinal_cols</code>, build the <code>ordinal_df</code> DataFrame. As you did in part three for the binary columns, apply the same process to the ordinal columns. That is, using the conversion tables provided at the beginning of the exercise, convert the values in <code>ordinal_df</code> to numbers.

```python
# TO-DO: create `ordinal_df` accordiing to above explanations
ordinal_df = pd.DataFrame()

education_order = ['illiterate', 'basic.4y', 'basic.6y', 'basic.9y', 'high.school',
                   'professional.course', 'university.degree', 'unknown']
day_of_week_order = ['mon', 'tue', 'wed', 'thu', 'fri']
month_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun',
               'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
poutcome_order = ['failure', 'nonexistent', 'success']

# فقط مقدارهای missing را با 'nonexistent' جایگزین کنید
df['poutcome'] = df['poutcome'].fillna('nonexistent')

ordinal_df['poutcome'] = pd.Categorical(df['poutcome'], categories=poutcome_order, ordered=True).codes
ordinal_df['education'] = pd.Categorical(df['education'], categories=education_order, ordered=True).codes
ordinal_df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=day_of_week_order, ordered=True).codes
ordinal_df['month'] = pd.Categorical(df['month'], categories=month_order, ordered=True).codes






ordinal_df.head()
```
